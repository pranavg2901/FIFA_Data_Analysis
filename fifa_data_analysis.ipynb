```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sc
```

```python
#nationality of players
fifa['nationality'].value_counts()
```

```python
# top 10 countries
fifa['nationality'].value_counts()[:10]
```

```python
# bar plot for top 10 countries player ratio
plt.figure(figsize=(8,8))
plt.bar(list(fifa['nationality'].value_counts()[:10].keys()),list(fifa['nationality'].value_counts()[:10]),color="g")
plt.show()
```

```python
player_salary=fifa[['short_name','wage_eur']]
```

```python
player_salary.head()
```

```python
#Asceding/descding order
player_salary=player_salary.sort_values(by=['wage_eur'],ascending=False)
player_salary.head()
```

```python
# bar plot for two values: short name and salary
plt.figure(figsize=(8,8))
plt.bar(list(player_salary['short_name'])[:5],list(player_salary['wage_eur'])[:5],color=["Blue","red","Green","Yellow","Pink"])
plt.show()
```
